```python
# Keywords to search for
#keywords = ["woman", "man"]
# keywords = ["beautiful"]
# keywords = ["dark"]
keywords = ["pretty","lovely","gorgeous"]

# Function to check and print elements containing keywords
def check_and_print_keywords(dataset, keywords):
    for index, element in enumerate(dataset):
        # Using any() to check if any keyword is in the element
        if any(keyword in element.lower() for keyword in keywords):
            print(f"Index: {index}, Element: '{element}'")
            
# Function to check and print elements containing keywords
def check_multiple_and_print_keywords(dataset, keywords):
    for index, element in enumerate(dataset):
        # Using any() to check if any keyword is in the element
        if all(keyword in element.lower() for keyword in keywords):
            print(f"Index: {index}, Element: '{element}'")

# Call the function
check_and_print_keywords(dataset['train']['prompt'], keywords)
```
